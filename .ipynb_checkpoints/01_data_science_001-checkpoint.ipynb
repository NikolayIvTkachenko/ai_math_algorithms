```python
i, n = symbols('i n')
summation = Sum(2 * i, (i, 1, n))
```

```python
# сложный процент
#
# p - первооначальный капитал
# r - годовая процентная ставка
# t - срок кредита (лет)
# n - периодичнеость начисления (кол-во месяцев в году)
#
# A - Остаток по кредиту
# P - первооначальный капитал
#
```

```python
p = 100
r = 0.20
t = 2
n = 12

a = p * (1 + (r / n)) ** (n * t)
print(p)
print(a)
```

```python
# вычисление пределов 
#
# оо - удвоенная латинская буква о - эт знак бесконечности
#
```

```python
from sympy import *
x = symbols('x')
f = 1 / x
result = limit(f, x, oo)
print(result)
```

```python
from sympy import *
n = symbols('n')
f = (1 + (1/n)) ** n
result = limit(f, n, oo)

print(result)
print(result.evalf())
```
