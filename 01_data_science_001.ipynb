```python
# Вычисление производной dz / dx исползованием цепноого правила и без него 
# приводит к одному и тому же результату
```

```python
from sympy import *
```

```python
x, y = symbols('x y')
```

```python
# Производная первоой функции
# Задаем имя с нижним подчеркиванием, чтобы не было коонфликта переменных
_y = x ** 2 + 2
```

```python
dy_dx = diff(_y)
```

```python
# Производная второй функции
# 
```

```python
z = y ** 3 - 2
```

```python
dz_dy = diff(z)
```

```python
# Вычисляем производную с помощью цепного правила
# и без него, подставляем функцйию y
#
```

```python
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))
```

```python
# Цепное  правило оработает
# оба варианта дают одинаковый результат
```

```python
print(dz_dx_chain)
print(dz_dx_no_chain)
```

```python
# Приближенное вычисление интекграла на Python
#
# f(x) = x ** 2 + 1 по пяти прямооугольникам от 0.0 до 1.0
#
```

```python
def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0
    
    for i in range(1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)
    
    return total_sum * delta_x
```

```python
def my_function(x):
    return x ** 2 + 1
```

```python
area = approximate_integral(a=0, b=1, n=5, f=my_function)
print(area)
```

```python
area = approximate_integral(a=0, b=1, n=1000, f=my_function)
print(area)
```

```python
area = approximate_integral(a=0, b=1, n=1000_000, f=my_function)
print(area)
```

```python
# Вычисляем интегралы с помощью пределов
#
#
```

```python
from sympy import *
```

```python
x, i, n = symbols('x i n')
```

```python
f = x ** 2 + 1
```

```python
lower, upper = 0, 1
```

```python
# Вычисляем ширину и высоту каждого пряямоугольника с индексом i
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)
```

```python
# Перебираем все n прямоугольников и суммируем их площади
n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()
```

```python
# Вычисляем площадь устремив число прямоугольников n к бесконечности
area = limit(n_rectangles, n, oo)
```

```python
print(area)
```

```python
# Медиана - это центральное значение в наборе упорядоченных значений 
# Если упорядочить значения по возрасьтанию, то медианой будет то значение, котоорое находится  середине этой последовательности.
```

```python
sample = [0, 1, 5, 7, 9, 10, 14]
```

```python
def median(values):
    ordered = sorted(values)
    n = len(ordered)
    mid = int(n / 2) - 1 if n % 2 ==0 else int(n / 2)
    
    if n % 2 == 0:
        return (ordered[mid]) + ordered[mid+1] / 2.0
    else:
        return ordered[mid]
```

```python
print(median(sample))
```

```python
# Медиана = это квантиль
# В описательной статистике есть понятие квантииелей. Квантиль, п сути, обобщает понятие медианы.
# Он разделяет данные не обязательно поосередине, а впроизвольной месте.
# Медиана - это фактически 50 %-ный квантиль, то есть значение, за которым находится 50 % упоряядоченных 
# значений. Медиана вместе с 25 и 75 %-нами квантилями называется квартилями, потоому ччто они разделдяют данные
# через каждые 25 %
```
